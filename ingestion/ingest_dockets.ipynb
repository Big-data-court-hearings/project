```python
"""
Bronze ingestion pipeline for docket data.
"""

from datetime import datetime
import logging

import pandas as pd

from api_client import fetch_paginated_data

from config import (
    DOCKETS_PATH,
    MAX_RECORDS,
)

# ============================================================
# LOGGING
# ============================================================

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)

logger = logging.getLogger(__name__)


def main() -> None:

    logger.info("Starting docket ingestion")

    # Create output directory
    DOCKETS_PATH.mkdir(parents=True, exist_ok=True)

    # Fetch API data
    records = fetch_paginated_data(
        endpoint="dockets/",
        max_records=MAX_RECORDS
    )

    # Convert to DataFrame
    df = pd.DataFrame(records)

    logger.info(f"DataFrame shape : {df.shape}")

    # Generate timestamped filename
    timestamp = datetime.now().strftime("%Y_%m_%d_%H%M")

    output_file = (
        DOCKETS_PATH /
        f"dockets_{timestamp}.parquet"
    )

    # Save parquet snapshot
    df.to_parquet(
        output_file,
        engine="pyarrow",
        index=False
    )

    logger.info(f"Parquet snapshot saved : {output_file}")


if __name__ == "__main__":
    main()
```
